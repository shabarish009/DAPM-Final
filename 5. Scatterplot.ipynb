```python
import matplotlib.pyplot as plt
import pandas as pd

# Load your dataset
file_path = r'C:\Users\asus\Desktop\Assignment\Second Assignment\Mod Global Economy Indicator 1.xlsx'  # Replace with your file path
data = pd.read_excel(file_path)

# Scatter plot of GDP vs. Sectoral Output
plt.figure(figsize=(10, 6))
plt.scatter(data['Gross Domestic Product (GDP)'], data['Sectoral Output'])
plt.title('Scatter plot of GDP vs. Sectoral Output')
plt.xlabel('Gross Domestic Product (GDP)')
plt.ylabel('Sectoral Output')
plt.grid(True)
plt.show()

# Scatter plot of GNI vs. Trade Balance
plt.figure(figsize=(10, 6))
plt.scatter(data['Gross National Income(GNI) in USD'], data['Trade Balance'])
plt.title('Scatter plot of GNI vs. Trade Balance')
plt.xlabel('Gross National Income (GNI) in USD')
plt.ylabel('Trade Balance')
plt.grid(True)
plt.show()

# Scatter plot of GDP vs. GNI
plt.figure(figsize=(10, 6))
plt.scatter(data['Gross Domestic Product (GDP)'], data['Gross National Income(GNI) in USD'])
plt.title('Scatter plot of GDP vs. GNI')
plt.xlabel('Gross Domestic Product (GDP)')
plt.ylabel('Gross National Income (GNI) in USD')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(data['Trade Balance'], data['Sectoral Output'])
plt.title('Trade Balance vs. Sectoral Output')
plt.xlabel('Trade Balance')
plt.ylabel('Sectoral Output')
plt.grid(True)
plt.show()
```
